# Seaborn

Create a numeric sample data


```python
import random 
#Explain 
from random import randint, seed
```

```python
# Generate a random list of 100 numbers

random_list = [randint(0,100) for i in range(100)]

# Generate the first 10 numbers of the list

random_list[:10]
```

```python
age = [randint(1,100) for i in range(100)]
weight = [randint(20,250) for i in range(100)]
height = [randint(1,8) for i in range(100)]

# Create a dataframe with the three lists
sample = pd.DataFrame({'Age': age, 'Weight': weight, 'Height': height})
```

```python
# Create a pair lot from seaborn
## kde diag_kind will drwa a kernel density estimate for each variable
sns.pairplot(sample, diag_kind = 'kde')
# sns.pairplot(sample, diag_kind = 'hist')
# sns.pairplot(sample, diag_kind = 'kde', kind = 'reg')

## Don't draw the diagonal plots
# sns.pairplot(sample, diag_kind = None)

```

Pair plot with condition

```python
sns.pairplot(sample, diag_kind = 'kde', hue = 'Height')
```

# 3D plot with plotly

```python
import plotly.graph_objects as go
```

```python

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=sample.Age,
    y=sample.Weight,
    z=sample.Height,
    mode='markers',
    marker=dict(
        size=4,
        color='blue',               
        opacity=0.8
    )
)])

# Set axes labels and title
fig.update_layout(scene=dict(
                    xaxis_title='Age',
                    yaxis_title='Weight',
                    zaxis_title='Height'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=30))

fig.show()
```
